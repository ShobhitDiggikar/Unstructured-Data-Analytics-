## Title: Text Processing Techniques**


```python
import pandas as pd
import numpy as np
import spacy
import nltk
```

```python
# en_core_web_lg is to a specific pre-trained spaCy model for the English language with a large vocabulary and word vectors.
#spaCy is an open-source natural language processing (NLP) library in Python.
#The lg in en_core_web_lg stands for "large," indicating that it is a larger and more comprehensive model compared to the smaller versions available.
import spacy.cli
spacy.cli.download("en_core_web_lg")
nlp = spacy.load("en_core_web_lg")
```

```python
reviews = pd.read_csv("/content/restaurant_reviews_az.csv")
reviews.head()
df=reviews
```

```python

reduced_reviews = df[(df['stars'] == 1) | (df['stars'] == 5)]
reduced_reviews.head(10)

```

```python
# Selection of the one star and five star reviews.
reduced_reviews = df[(df['stars'] == 1) | (df['stars'] == 5)]

# Taking only first 100 rows of selected data.
reduced_reviews_subset = reduced_reviews.head(1000)

# Printing first 10 rows of the selected subset
reduced_reviews_subset.head(10)
```

Tokenization with SpaCy

```python
# Applying tokenization using SpaCy.
reduced_reviews_subset['tokenized_text'] = reduced_reviews_subset['text'].apply(lambda x: [token.text for token in nlp(x)])

# Displaying the tokenized text for first 10 reviews.
for i in range(10):
    print(f"Review {i + 1} tokens: {reduced_reviews_subset['tokenized_text'].iloc[i]}")
```

Tokenization with NLTK

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
reduced_reviews_subset['tokenized_text_nltk'] = reduced_reviews_subset['text'].apply(lambda x: word_tokenize(x))
for i in range(10):
    print(f"Review {i + 1} tokens (NLTK): {reduced_reviews_subset['tokenized_text_nltk'].iloc[i]}")
```

Part of speech tagging


```python
# Apply POS tagging using SpaCy
reduced_reviews_subset['pos_tags_spacy'] = reduced_reviews_subset['text'].apply(lambda x: [(token.text, token.pos_) for token in nlp(x)])

# Display the POS tags for the first 10 reviews
for i in range(10):
    print(f"Review {i + 1} POS tags (SpaCy): {reduced_reviews_subset['pos_tags_spacy'].iloc[i]}")
```

```python
nltk.download('averaged_perceptron_tagger')

reduced_reviews_subset['pos_tags_nltk'] = reduced_reviews_subset['tokenized_text_nltk'].apply(lambda x: nltk.pos_tag(x))
for i in range(10):
    print(f"Review {i + 1} POS tags (NLTK): {reduced_reviews_subset['pos_tags_nltk'].iloc[i]}")
```

STEMMING

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Create a Porter stemmer instance
porter_stemmer = PorterStemmer()

# Apply stemming using NLTK
reduced_reviews_subset['stemmed_text_nltk'] = reduced_reviews_subset['tokenized_text_nltk'].apply(lambda x: [porter_stemmer.stem(word) for word in x])

# Display the stemmed text for the first 10 reviews
for i in range(10):
    print(f"Review {i + 1} stemmed text (NLTK): {reduced_reviews_subset['stemmed_text_nltk'].iloc[i]}")
```

Lemmatization

```python
# Application of lemmatization using SpaCy
reduced_reviews_subset['lemmatized_text_spacy'] = reduced_reviews_subset['text'].apply(lambda x: [token.lemma_ for token in nlp(x)])

# Displaying the lemmatized text for the first 10 reviews
for i in range(10):
    print(f"Review {i + 1} lemmatized text (SpaCy): {reduced_reviews_subset['lemmatized_text_spacy'].iloc[i]}")
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

# Download NLTK resources if not already downloaded
nltk.download('wordnet')

# Create a WordNet lemmatizer instance
wordnet_lemmatizer = WordNetLemmatizer()

# Apply lemmatization using NLTK
reduced_reviews_subset['lemmatized_text_nltk'] = reduced_reviews_subset['tokenized_text_nltk'].apply(lambda x: [wordnet_lemmatizer.lemmatize(word, pos='v') for word in x])

# Display the lemmatized text for the first 10 reviews
for i in range(10):
    print(f"Review {i + 1} lemmatized text (NLTK): {reduced_reviews_subset['lemmatized_text_nltk'].iloc[i]}")
```

```python
reduced_reviews_subset['entities_spacy'] = reduced_reviews_subset['text'].apply(lambda x: [(ent.text, ent.label_) for ent in nlp(x).ents])
for i in range(10):
    print(f"Review {i + 1} entities (SpaCy): {reduced_reviews_subset['entities_spacy'].iloc[i]}")
```

```python
from nltk import ne_chunk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('maxent_ne_chunker')
nltk.download('words')

# Apply POS tagging using NLTK (if not done already)
reduced_reviews_subset['pos_tags_nltk'] = reduced_reviews_subset['text'].apply(lambda x: pos_tag(word_tokenize(x)))

# Apply Named Entity Recognition using NLTK
reduced_reviews_subset['entities_nltk'] = reduced_reviews_subset['pos_tags_nltk'].apply(lambda x: ne_chunk(x))

# Display the entities for the first 10 reviews
for i in range(10):
    print(f"Review {i + 1} entities (NLTK): {reduced_reviews_subset['entities_nltk'].iloc[i]}")
```

Top 20vNouns

```python
from collections import Counter

def extract_nouns(text):
    doc = nlp(text)
    return [token.text.lower() for token in doc if token.pos_ == 'NOUN']

one_star_nouns_subset = reduced_reviews_subset[reduced_reviews_subset['stars'] == 1]['text'].apply(extract_nouns)
one_star_nouns_combined_subset = [noun for sublist in one_star_nouns_subset for noun in sublist]

one_star_nouns_counts_subset = Counter(one_star_nouns_combined_subset)
top_20_one_star_nouns_subset = one_star_nouns_counts_subset.most_common(20)
print("Top 20 Nouns in 1-star reviews (Subset):", top_20_one_star_nouns_subset)

five_star_nouns_subset = reduced_reviews_subset[reduced_reviews_subset['stars'] == 5]['text'].apply(extract_nouns)
five_star_nouns_combined_subset = [noun for sublist in five_star_nouns_subset for noun in sublist]

five_star_nouns_counts_subset = Counter(five_star_nouns_combined_subset)

# Get the top 20 nouns
top_20_five_star_nouns_subset = five_star_nouns_counts_subset.most_common(20)
print("Top 20 Nouns in 5-star reviews (Subset):", top_20_five_star_nouns_subset)
```

**Top 20 Adjecectives**

```python
from collections import Counter

# Define a function to extract adjectives from the text
def extract_adjectives(text):
    doc = nlp(text)
    return [token.text.lower() for token in doc if token.pos_ == 'ADJ']

# Apply the function to 1-star reviews in reduced_reviews_subset
one_star_adjectives_subset = reduced_reviews_subset.loc[reduced_reviews_subset['stars'] == 1, 'text'].apply(extract_adjectives)

# Combine the lists of adjectives into a single list
one_star_adjectives_combined_subset = [adj for sublist in one_star_adjectives_subset for adj in sublist]

# Count the occurrences of each adjective
one_star_adjectives_counts_subset = Counter(one_star_adjectives_combined_subset)

# Get the top 20 adjectives
top_20_one_star_adjectives_subset = one_star_adjectives_counts_subset.most_common(20)
print("Top 20 Adjectives in 1-star reviews (Subset):", top_20_one_star_adjectives_subset)

# Apply the function to 5-star reviews in reduced_reviews_subset
five_star_adjectives_subset = reduced_reviews_subset.loc[reduced_reviews_subset['stars'] == 5, 'text'].apply(extract_adjectives)

# Combine the lists of adjectives into a single list
five_star_adjectives_combined_subset = [adj for sublist in five_star_adjectives_subset for adj in sublist]

# Count the occurrences of each adjective
five_star_adjectives_counts_subset = Counter(five_star_adjectives_combined_subset)

# Get the top 20 adjectives
top_20_five_star_adjectives_subset = five_star_adjectives_counts_subset.most_common(20)
print("Top 20 Adjectives in 5-star reviews (Subset):", top_20_five_star_adjectives_subset)
```

Top 20 Verbs

```python
from collections import Counter

# Define a function to extract verbs from the text
def extract_verbs(text):
    doc = nlp(text)
    return [token.text.lower() for token in doc if token.pos_ == 'VERB']

# Apply the function to 1-star reviews in reduced_reviews_subset
one_star_verbs_subset = reduced_reviews_subset.loc[reduced_reviews_subset['stars'] == 1, 'text'].apply(extract_verbs)

# Combine the lists of verbs into a single list
one_star_verbs_combined_subset = [verb for sublist in one_star_verbs_subset for verb in sublist]

# Count the occurrences of each verb
one_star_verbs_counts_subset = Counter(one_star_verbs_combined_subset)

# Get the top 20 verbs
top_20_one_star_verbs_subset = one_star_verbs_counts_subset.most_common(20)
print("Top 20 Verbs in 1-star reviews (Subset):", top_20_one_star_verbs_subset)

# Apply the function to 5-star reviews in reduced_reviews_subset
five_star_verbs_subset = reduced_reviews_subset.loc[reduced_reviews_subset['stars'] == 5, 'text'].apply(extract_verbs)

# Combine the lists of verbs into a single list
five_star_verbs_combined_subset = [verb for sublist in five_star_verbs_subset for verb in sublist]

# Count the occurrences of each verb
five_star_verbs_counts_subset = Counter(five_star_verbs_combined_subset)

# Get the top 20 verbs
top_20_five_star_verbs_subset = five_star_verbs_counts_subset.most_common(20)
print("Top 20 Verbs in 5-star reviews (Subset):", top_20_five_star_verbs_subset)
```

Top 20 most often used named


```python
from collections import Counter

# Define a function to extract named entities from the text
def extract_named_entities(text):
    doc = nlp(text)
    return [ent.text for ent in doc.ents]

# Apply the function to all reviews in reduced_reviews_subset
all_named_entities_subset = reduced_reviews_subset['text'].apply(lambda x: extract_named_entities(x))

# Combine the lists of named entities into a single list
all_named_entities_combined_subset = [entity for sublist in all_named_entities_subset for entity in sublist]

# Count the occurrences of each named entity
named_entities_counts_subset = Counter(all_named_entities_combined_subset)

# Get the top 20 named entities
top_20_named_entities_subset = named_entities_counts_subset.most_common(20)
print("Top 20 Named Entities in reviews (Subset):", top_20_named_entities_subset)
```

The primary noun prevalent in both 1-star and 5-star reviews is "food," which is unsurprising. Additionally, commonly used words like "time" and "service" are observed. Notably, terms such as 'Minutes' and 'Manager' frequently occur in 1-star reviews, suggesting prolonged wait times and instances where customers had to escalate issues to the manager. Adjectives like "worst," "bad," "terrible," and "rude" are frequently used in 1-star reviews, indicating negative experiences. In contrast, 5-star reviews feature positive descriptors such as "amazing," "great," "good," and "nice." This stark contrast in language makes it easy to distinguish between negative and positive reviews based on the expressions used.
